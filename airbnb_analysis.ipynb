## Import Libraries

```python
import pandas as pd
```

## Load Data

```python
df = pd.read_csv('AB_US_2023.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['last_review'] = pd.to_datetime(df['last_review'])
```

```python
df = df.rename(columns = {'neighbourhood_group':'neighborhood_group','neighbourhood':'neighborhood'})
print(df['neighborhood_group'].unique())
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df['listing_id'] = df.index
```

```python
df.head()
```

## Create Schema Tables

**Dimension tables**

```python
# Host dimension
host_dim = df[['host_id', 'host_name']].reset_index(drop=True)
host_dim['host_dim_id'] = host_dim.index
host_dim = host_dim[['host_dim_id', 'host_id','host_name']]
host_dim.head()
```

```python
# Neighborhood dimension 
neighborhood_dim = df[['neighborhood', 'neighborhood_group']].reset_index(drop=True)
neighborhood_dim['neighborhood_id'] = neighborhood_dim.index
neighborhood_dim = neighborhood_dim[['neighborhood_id', 'neighborhood', 'neighborhood_group']]
```

```python
# Location dimension
location_dim = df[['latitude', 'longitude']].drop_duplicates().reset_index(drop=True)
location_dim['location_id'] = location_dim.index
location_dim = location_dim[['location_id', 'latitude', 'longitude']]
```

```python
# Reviews dimension
reviews_dim = df[['number_of_reviews', 'last_review', 'reviews_per_month', 'number_of_reviews_ltm']].reset_index(drop=True)
reviews_dim['last_review'] = reviews_dim['last_review']
reviews_dim['last_review_day'] = reviews_dim['last_review'].dt.day
reviews_dim['last_review_month'] = reviews_dim['last_review'].dt.month
reviews_dim['last_review_year'] = reviews_dim['last_review'].dt.year
reviews_dim['reviews_id'] = reviews_dim.index

reviews_dim.head()
```
